```python
USERS_EXPORT_FILE="../source_data/users.csv"
CONTENT_EXPORT_FILE="../source_data/content.csv"
CONTENT_USERS_RANGE=(50,400)
```

```python
# Load Users from csv
# Load current users
import csv
users = []
with open(USERS_EXPORT_FILE, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        users.append(row)
users
```

```python
# Convert to datetime dates
import datetime
today = datetime.datetime.today()
for user in users:
    user["ValidFrom"] = datetime.datetime.strptime(user["ValidFrom"], "%Y-%m-%d")
    user["ValidTo"] = datetime.datetime.strptime(user["ValidTo"], "%Y-%m-%d") if user["ValidTo"] != '2999-12-31' else today
users
```

```python
# Load Users from csv
# Load current content
import csv
content = []
with open(CONTENT_EXPORT_FILE, newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        content.append(row)
content
```

```python
# Convert to datetime dates
import datetime
today = datetime.datetime.today()
for c in content:
    c["StartDate"] = datetime.datetime.strptime(c["StartDate"], "%Y-%m-%d %H:%M:%S")
    c["EndDate"] = datetime.datetime.strptime(c["EndDate"], "%Y-%m-%d %H:%M:%S") if c["EndDate"] != '2999-12-31' else today
users
```
